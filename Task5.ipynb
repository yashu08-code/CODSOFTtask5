```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('creditcard.csv')
df
```

```python
df.info()
```

```python
#check the number of missing values
df.isnull().sum()
```

```python
#check distrubution of legit and fraud transaction
df['Class'].value_counts()
```

```python
legit = df[df.Class==0]
fraud = df[df.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare values of both transaction 
df.groupby('Class').mean()
```

```python
#under sampling 
#build a sample dataset containing 
#no. of fraudulant transactions == 492
legit_sample = legit.sample(n=492)
```

```python
#concatenate 2 data frames
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#split the data into features and targets
x = new_dataset.drop(columns='Class',axis=1)
y = new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
#split data into training and testing data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
#Model training = Logistic Regression 
model = LogisticRegression()
```

```python
#training the model with training model
model.fit(x_train,y_train)
```

```python
#Model evaluation based on accuracy score
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy on trainig data: ' , training_data_accuracy)
```

```python
#accuracy on test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy on testing data: ' , test_data_accuracy)
```

```python
# Make predictions on the test data
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import precision_score
precision = precision_score(y_test, y_pred)
print(f"Precision: {precision:.2f}")
```

```python
from sklearn.metrics import recall_score
recall = recall_score(y_test, y_pred)
print(f"Recall: {recall:.2f}")
```

```python
from sklearn.metrics import f1_score
f1 = f1_score(y_test, y_pred)
print(f"F1 Score: {f1:.2f}")
```
